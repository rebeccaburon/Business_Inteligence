# Create a correlation matrix and a heat map of all data and investigate it. Tell which wine attribute has the biggest influence on the wine quality. Which has the lowest? Are there any attributes, apart from the wine quality, which are highly correlated?

```python
#Create correlation matrix
correlation_matrix = df_all.corr(numeric_only=True)

```

```python
print(correlation_matrix)
```

#### Create Heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Wine Attributes")
plt.show()
```

```python
# Unstack, sort and filter out weak correlations

high_correlation_pairs = (
    correlation_matrix
    .abs()
    .unstack()
    .sort_values(ascending=False)
    .drop_duplicates()
)
```

```python
# Removes self-correlations, any rows where the first variable in the pair is quality and assuring only strong correlation above 0.5.

high_correlation_pairs = high_correlation_pairs[
    (high_correlation_pairs < 1.0) &
    (~high_correlation_pairs.index.get_level_values(0).isin(['quality'])) & ## ~ = (not)
    (~high_correlation_pairs.index.get_level_values(1).isin(['quality'])) &
    (high_correlation_pairs > 0.5)
]
```

```python
high_correlation_pairs = high_correlation_pairs.to_frame(name='Correlation (r)')
print("Highly correlated attribute pairs (r > 0.5):")
print(high_correlation_pairs)
```
